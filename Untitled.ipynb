```python
strain = f['strain']['Strain'].value
```

```python
ts = f['strain']['Strain'].attrs['Xspacing']#xspacing gives time between each sample
```

```python
ts
```

```python
metaKeys = f['meta'].keys()
```

```python
metaKeys
```

```python
meta = f['meta']
```

```python
meta
```

```python
for key in metaKeys:
    print (key, meta[key].value)
```

```python
gps=meta["GPSstart"].value
duration=meta["Duration"].value
gpsend=gps+duration
time=np.arange(gps,gpsend,ts)
```

```python
samples=10000
plt.plot(time[0:samples],strain[0:samples])
plt.xlabel("GPS")
plt.ylabel("strain")
plt.title("Strain data")#plotting ligo data for low band 20hz frequency
```
